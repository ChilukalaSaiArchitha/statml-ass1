```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis, mode

# Create the DataFrame
data = [
    {"Name": "Hasini", "CGPA": 8.64},
    {"Name": "Sahithya", "CGPA": 9.09},
    {"Name": "Shruthi", "CGPA": 9.24},
    {"Name": "Deepthi", "CGPA": 9.45},
    {"Name": "Ronith", "CGPA": 6.3},
    {"Name": "Harshini", "CGPA": 8.5},
    {"Name": "Architha", "CGPA": 7.32},
    {"Name": "Varshitha", "CGPA": 9.4},
    {"Name": "Varshith", "CGPA": 9.0},
    {"Name": "Vishal", "CGPA": 8.5},
    {"Name": "Sri Kumar", "CGPA": 7.5},
    {"Name": "Preethi", "CGPA": 8.5},
    {"Name": "Akshaya", "CGPA": 9.0},
    {"Name": "Madhavi", "CGPA": 8.0},
    {"Name": "Srinidhi", "CGPA": 8.0},
    {"Name": "Abhilash", "CGPA": 9.3},
    {"Name": "Hariharan", "CGPA": 7.4},
    {"Name": "Bal Reddy", "CGPA": 7.3},
    {"Name": "Rahul", "CGPA": 8.0},
    {"Name": "Ananya", "CGPA": 7.31}
]

df = pd.DataFrame(data)

# Calculate required statistics
mean_cgpa = df['CGPA'].mean()
median_cgpa = df['CGPA'].median()

# Correct way to calculate mode
mode_result = mode(df['CGPA'], keepdims=True)
mode_cgpa = mode_result.mode[0] if mode_result.count[0] > 1 else df['CGPA'].mode()[0]

variance_cgpa = df['CGPA'].var()
std_dev_cgpa = df['CGPA'].std()
range_cgpa = df['CGPA'].max() - df['CGPA'].min()
q1 = df['CGPA'].quantile(0.25)
q3 = df['CGPA'].quantile(0.75)
iqr_cgpa = q3 - q1
skewness_cgpa = skew(df['CGPA'])
kurtosis_cgpa = kurtosis(df['CGPA'])

# Print the results
print(f"Mean CGPA: {mean_cgpa}")
print(f"Median CGPA: {median_cgpa}")
print(f"Mode CGPA: {mode_cgpa}")
print(f"Variance CGPA: {variance_cgpa}")
print(f"Standard Deviation CGPA: {std_dev_cgpa}")
print(f"Range CGPA: {range_cgpa}")
print(f"Interquartile Range (IQR) CGPA: {iqr_cgpa}")
print(f"Skewness CGPA: {skewness_cgpa}")
print(f"Kurtosis CGPA: {kurtosis_cgpa}")
```
